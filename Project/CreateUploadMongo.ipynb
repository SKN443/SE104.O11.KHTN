```python
import glob
import cv2
import json
import io
from PIL import Image

dct = []
emb = {}
with open('/content/database_tmp.json') as f:
    tmps = json.load(f)
    for tmp in tmps:
        emb[tmp['metadata']['image_path']] = tmp
for img_path in glob.glob('/content/imgdata/*.jpg'):
    img = Image.open(img_path)

    image_bytes = io.BytesIO()
    img.save(image_bytes, format='JPEG')

    dct.append({
        'image': image_bytes.getvalue(),
        'path': img_path[len('/content/'):],
        'vector': emb[img_path[len('/content/'):]]['vector'],
        'product_id': emb[img_path[len('/content/'):]]['metadata']['product_id'],
        'category': emb[img_path[len('/content/'):]]['metadata']['category'],
        'flag' : 1
    })
```

```python
db.database.insert_many(dct)
```
